```python
import pandas as pd
import os
from sql_table_schema_parser import SQLTableSchemaParser 
from sql_commands import create_table
```

```python
# Instantiate parser
sql_parser = SQLTableSchemaParser(sql_command=create_table)

# Get schema and date columns
schema_dict = sql_parser.get_schema_dict()
indices_by_type = sql_parser.get_indices_by_type()
date_col_indices = sql_parser.get_date_col_indices()

columns = schema_dict.keys()

bool_col_indices = sql_parser.get_bool_col_indices()
```

```python
test_file = '/Users/matthewtryba/Library/CloudStorage/Dropbox/01 Matt/Tryba Documents/GitHub/Portfolio-Projects/05 Mortgage Home Loan Analysis/test_csv_folder/2023_sample.csv'
df = pd.read_csv(test_file, sep="|", header=None, names=columns)
df.head()

for key, value in indices_by_type.items():
    if key == 'DATE':
        for i in value:
            df.iloc[:,i].astype('datetime64[ns]')
    elif key == 'BOOLEAN':
        for i in value:
            df.iloc[:,i].astype('bool')
    elif key == "INTEGER":
        for i in value:
            df.iloc[:,i].astype('Int64')
    elif key == "NUMERIC":
        for i in value:
            df.iloc[:,i].astype('float')
    elif key == 'VARCHAR':
        for i in value:
            df.iloc[:,i].astype('object')
```
